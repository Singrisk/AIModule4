```python
import numpy as np

def svd_manual(A):
    # Step 1: A^T A
    ATA = A.T @ A
    
    # Step 2: Eigen decomposition
    eigenvalues, V = np.linalg.eig(ATA)
    
    # Step 3: Sort eigenvalues and vectors
    idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[idx]
    V = V[:, idx]

    # Step 4: Singular values
    singular_values = np.sqrt(eigenvalues)

    # Step 5: Compute U
    U = []
    for i in range(len(singular_values)):
        if singular_values[i] > 1e-10:
            u = (A @ V[:, i]) / singular_values[i]
            U.append(u)
    U = np.array(U).T

    # Sigma matrix
    Sigma = np.zeros((U.shape[1], V.shape[1]))
    np.fill_diagonal(Sigma, singular_values[:Sigma.shape[0]])

    return U, Sigma, V.T
```

```python
import numpy as np

# Example data
A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

# Your SVD
U, Sigma, VT = svd_manual(A)

# Choose reduced dimension
k = 1

# Reduce
U_k = U[:, :k]
Sigma_k = Sigma[:k, :k]
VT_k = VT[:k, :]

# Reduced data
A_reduced = U_k @ Sigma_k

# Reconstruction
A_approx = U_k @ Sigma_k @ VT_k

print("Reduced Data:\n", A_reduced)
print("\nReconstructed Matrix:\n", A_approx)
```
