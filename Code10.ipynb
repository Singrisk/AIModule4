# Using Scikit-learn

```python
# Import necessary Libraries. 
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Declare Dataset
data = {
    "x": [1, 2, 3, 8, 9, 10],
    "y": [2, 3, 4, 7, 8, 9]
}

df = pd.DataFrame(data)
```

```python
# Defining and training model. 
kmeans = KMeans(
    n_clusters=2, 
    random_state=42
)

df["cluster"] = kmeans.fit_predict(df)
```

```python
# Print Dataset. 
print(df)
```

```python
# Printing cluster centers. 
print(kmeans.cluster_centers_)
```
