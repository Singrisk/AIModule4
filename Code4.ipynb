## Python Implementation (Normal Equation — From Scratch)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("boston_house_prices.csv", header = 1) 

df.rename(columns={df.columns[-1]: "target"}, inplace=True) 

X = df.drop("target", axis=1) 
y = df["target"] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42) 

scaler = StandardScaler() 
X_train=scaler.fit_transform(X_train) 
X_test = scaler.transform(X_test)
```

```python
# add bias column to TRAIN and TEST
X_train_b = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_b  = np.c_[np.ones(X_test.shape[0]), X_test]
```

```python
# normal equation using TRAIN data only
beta = np.linalg.inv(X_train_b.T @ X_train_b) @ X_train_b.T @ y_train

print("Intercept:", beta[0])
print("Weights:", beta[1:])
```

```python
y_pred = X_test_b @ beta
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

## Python Implementation (Gradient Descent — Scratch)

```python
df = pd.read_csv("boston_house_prices.csv", header=1)
df.rename(columns={df.columns[-1]: "target"}, inplace=True)

X = df.drop("target", axis=1)
y = df["target"].values   # convert to numpy

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.30, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
X_train_b = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_b  = np.c_[np.ones(X_test.shape[0]), X_test]
```

```python
# parameters
lr = 0.01
epochs = 5000
n = X_train_b.shape[0]
```

```python
# initialize weights
beta = np.zeros(X_train_b.shape[1])

for i in range(epochs):

    # prediction
    y_pred = X_train_b @ beta

    # gradient (vectorized)
    gradient = (2/n) * (X_train_b.T @ (y_pred - y_train))

    # update rule
    beta -= lr * gradient

    # optional: print loss
    if i % 500 == 0:
        loss = np.mean((y_pred - y_train)**2)
        print(f"Epoch {i} Loss: {loss:.4f}")
```

```python
y_pred = X_test_b @ beta
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
