## Python Implementation (Normal Equation — From Scratch)

```python
import numpy as np
```

```python
# dataset: experience vs salary
X = np.array([1,2,3,4,5])
y = np.array([15,18,25,32,38])
```

```python
# add bias column (1s)
X = np.c_[np.ones(len(X)), X]
```

```python
# normal equation
beta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Intercept:", beta[0])
print("Slope:", beta[1])
```

```python
# prediction
def predict(x):
    return beta[0] + beta[1]*x

print("Salary for 6 years:", predict(6))
```

## Python Implementation (Gradient Descent — Scratch)

```python
# data
X = np.array([1,2,3,4,5])
y = np.array([15,18,25,32,38])
```

```python
# add bias
X = np.c_[np.ones(len(X)), X]
```

```python
# parameters
beta = np.zeros(2)
lr = 0.01
epochs = 1000
n = len(y)
```

```python
for _ in range(epochs):
    y_pred = X @ beta
    gradient = -(2/n) * (X.T @ (y - y_pred))
    beta = beta - lr * gradient

print("Intercept:", beta[0])
print("Slope:", beta[1])
```

```python
print("Prediction for 6:", beta[0] + beta[1]*6)
```
