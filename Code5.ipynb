## Python Implementation (Normal Equation — From Scratch)

```python
import numpy as np

# data (quadratic)
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
# degree
d = 2
```

```python
# build polynomial features
X = np.column_stack([x**i for i in range(d+1)])
```

```python
# normal equation
beta = np.linalg.pinv(X) @ y

print("Coefficients:", beta)
```

```python
def predict(val):
    return sum(beta[i]*(val**i) for i in range(len(beta)))

print("Prediction for 6:", predict(6))
```

## Python Implementation (Gradient Descent — Scratch)

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
degree = 2
```

```python
# scale FIRST
x_mean = np.mean(x)
x_std = np.std(x)
z = (x - x_mean)/x_std
```

```python
# polynomial features from scaled variable
X = np.column_stack([z**i for i in range(degree+1)])
```

```python
beta = np.zeros(degree+1)
lr = 0.1
n = len(y)
```

```python
for _ in range(5000):
    y_pred = X @ beta
    gradient = -(2/n)*(X.T @ (y - y_pred))
    beta -= lr * gradient
```

```python
print("beta:", beta)
```

```python
# ---- prediction ----
def predict(val):
    z = (val - x_mean)/x_std
    features = np.array([z**i for i in range(degree+1)])
    return features @ beta

print("Prediction for 6:", predict(6))
```
